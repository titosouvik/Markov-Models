```python
import numpy as np
import matplotlib.pyplot as plt
import string
from sklearn.model_selection import train_test_split
```

```python
input_files = ['data/edgar_allan_poe.txt','data/robert_frost.txt']
```

```python
# Collect data

from tkinter.filedialog import Open


input_texts = []
labels = []

for label, f in enumerate(input_files):
    print(f"{f} corresponds to label {label}")

    for line in open(f):
        line = line.rstrip().lower()
        if line:
            ## Rremove special characters
            line = line.translate(str.maketrans('','',string.punctuation))

            input_texts.append(line)
            labels.append(label)

```

```python
train_text, test_text, y_train, y_test = train_test_split(input_texts, labels)
```

```python
## Convert text to integers

idx = 1

word2idx = {'<unk>':0}

for text in train_text:
    tokens = text.split()

    for token in tokens:
        if token not in word2idx:
            word2idx[token] = idx
            idx += 1
            
```

```python
# Convert text data to integers

train_text_int = []
test_text_int = []


for text in train_text:
    tokens = text.split()
    line_int = [word2idx[token] for token in tokens]
    train_text_int.append(line_int)

for text in test_text:
    tokens = text.split()
    line_int = [word2idx.get(token,0) for token in tokens]
    test_text_int.append(line_int)
```

```python
## Initialize transtion matrix (A) and intial distribution function (Pi)

v = len(word2idx)

A0 = np.ones((v,v))
Pi0 = np.ones(v)

A1 = np.ones((v,v))
Pi1 = np.ones(v)
```

```python
def compute_counts(text_ints, A, pi):
    for token in text_ints:
        last_idx = None
        for idx in token:
            if last_idx is None: # For first word
                pi[idx] +=1
            else:
                A[last_idx,idx] +=1 # Count transition

            idx = last_idx

            
```

```python
compute_counts([t for t,y in zip(train_text_int, y_train) if y==0],A0,Pi0)
compute_counts([t for t,y in zip(train_text_int, y_train) if y==1],A1,Pi1)
```

```python
## Normalize the A and Pi

A0 /= A0.sum(axis=1,keepdims=True)
Pi0 /= Pi0.sum()


A1 /= A1.sum(axis=1,keepdims=True)
Pi1 /= Pi1.sum()
```

```python
## Compute Log Probabilities

logA0 = np.log10(A0)
logPi0 = np.log10(Pi0)


logA1 = np.log10(A1)
logPi1 = np.log10(Pi1)
```

```python
## Compute Priors

count0 = sum(y==0 for y in y_train)
count1 = sum(y==1 for y in y_train)

p0 = count0/len(y_train)
p1 = count1/len(y_train)

logp0 = np.log10(p0)
logp1 = np.log(p1)

print(p0,p1)
```

```python
## Classifier

class MarkovClassifier:
    def __init__(self, logAs, logPis,logpriors):
        self.logAs = logAs
        self.logPis = logPis
        self.logpriors = logpriors

        self.k = len(logpriors) # number of classes

    def compute_log_likelihood(self, input_,class_):
        logA = self.logAs[class_]
        logPi = self.logPis[class_]

        last_idx = None
        logprob = 0
        for idx in input_:
            if last_idx is None: # For first word
                logprob += logPi[idx]
            else:
                logprob += logA[last_idx,idx] # Count transition

            idx = last_idx

        return logprob

    def predict(self, inputs):
        predictions = np.zeros(len(inputs))
        for i, input_ in enumerate(inputs):
            posterior = [self.compute_log_likelihood(input_, c) + self.logpriors[c] for c in range(self.k)]

            pred = np.argmax(posterior)

            predictions[i] = pred

        return predictions
```

```python
clf = MarkovClassifier([logA0,logA1],[logPi0,logPi1],[logp0,logp1])
```

```python
Ptrain = clf.predict(train_text_int)
```

```python
print(f"Accuracy {np.mean(Ptrain==y_train)}")
```

```python
Ptest = clf.predict(test_text_int)
print(f"Accuracy {np.mean(Ptest==y_test)}")
```
